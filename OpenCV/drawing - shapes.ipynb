```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
blank = np.zeros((512, 512, 3), dtype=np.int16)
plt.imshow(blank)
```

```python
rect = cv2.rectangle(blank, pt1=(100,100), pt2=(400,400), color=(0,255,0),thickness=10)
plt.imshow(blank)
```

```python
circle1 = cv2.circle(blank, center=(400,400), radius=50,color=(255,0,0),thickness=4)
circle  = cv2.circle(blank, center=(100,100), radius=50,color=(255,0,0),thickness=-1)
plt.imshow(blank)
```

```python
line = cv2.line(blank, (0, 0), (512, 512), color=(0,124, 125), thickness=2)
plt.imshow(blank)
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
text = cv2.putText(blank, 'hello', (10, 500), fontFace=font, fontScale=4, color=(255,255,255), thickness=3,
                  lineType=cv2.LINE_AA)
plt.imshow(blank)
```
