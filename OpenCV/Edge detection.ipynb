```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# Canny Edge detection

```python
img = cv2.imread('data/sammy_face.jpg')
plt.imshow(img)
```

```python
edges = cv2.Canny(img, threshold1=127, threshold2=127)
plt.imshow(edges)
```

```python
median = np.median(img)
median
```

```python
th1 = int(max(0,0.7*median)) # lower = 0 or 70% of median value
th2 = int(min(255, 1.3*median)) # upper = 255 or 30% above the median

th1, th2
```

```python
edges = cv2.Canny(img, threshold1=th1, threshold2=th2)
plt.imshow(edges)
```

```python
blurred = cv2.blur(img, ksize=(8,8))
edges = cv2.Canny(blurred, threshold1=th1, threshold2=th2)
plt.imshow(edges)
```
