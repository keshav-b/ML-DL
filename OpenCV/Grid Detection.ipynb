```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('data/flat_chessboard.png')
plt.imshow(img)
```

```python
found, grid = cv2.findChessboardCorners(img, (5,5))
```

```python
found
```

```python
cv2.drawChessboardCorners(img,(5,5), grid, found)
plt.imshow(img)
```

```python
img = cv2.imread('data/dot_grid.png')
plt.imshow(img)
```

```python
found, grid = cv2.findCirclesGrid(img, (5,5), cv2.CALIB_CB_SYMMETRIC_GRID)
```

```python
cv2.drawChessboardCorners(img,(5,5), grid, found)
plt.imshow(img)
```
