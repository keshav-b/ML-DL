```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# Harris Corner Detection

"Corners can be identified by looking at the significant change in all directions."

```python
img = cv2.imread('data/flat_chessboard.png')
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

gray = np.float32(gray)
dst = cv2.cornerHarris(src=gray,blockSize=2,ksize=3,k=0.04)

dst = cv2.dilate(dst,None)
img[dst>0.01*dst.max()]=[255,0,0]

plt.imshow(img, cmap='gray')
```

# Shi Tomasi Corner Detection


```python
img = cv2.imread('data/flat_chessboard.png')
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

gray = np.float32(gray)

corners = cv2.goodFeaturesToTrack(image=gray,maxCorners=100,qualityLevel=0.01,minDistance=10)
corners = np.int0(corners)

for i in corners:
    x,y = i.ravel() # flatten array
    cv2.circle(img,(x,y),3,255,-1)

plt.imshow(img)
```
