```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('data/internal_external.png', 0)
plt.imshow(img, cmap='gray')
```

```python
# internal and external countours
contours, heirarchy = cv2.findContours(img, cv2.RETR_CCOMP, cv2.CHAIN_APPROX_SIMPLE)
```

```python
external = np.zeros(img.shape)
internal = np.zeros(img.shape)
```

```python
for i in range(len(contours)):
    if heirarchy[0][i][3] == -1: # checking out last column => external contour
        cv2.drawContours(external, contours, i, 255, -1)
    else:
        cv2.drawContours(internal, contours, i, 255, -1)
        
        
        
```

```python
plt.imshow(external, cmap='gray')
```
